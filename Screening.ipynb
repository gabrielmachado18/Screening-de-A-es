## **3. Screening**
Inicialmente, faremos a triagem das **50 primeiras** empresas com base nos seguintes parâmetros:

* **Patrimônio Líquido (patrliq)** positivo;

* **Dívida Bruta sobre Patrimônio Líquido (divbpatr)** não pode ser maior que 30%

A base de dados da **Fundamentus** não remove ações de empresas que **não estão mais operando**. Dessa forma, usaremos a função `is_traded_recently` para avaliar se **houve negócios nos últimos 5 dias das empresas selecionadas**.

Após a aplicação da função `is_traded_recently`, escolheremos as **10 primeiras empresas** que apresentarem os **maiores ROE** (*return on equity*)

### **3.1 Aplicando os Parâmetros** 

```python
stocks = df[(df.patrliq > 0) & (df.divbpatr <= 0.6)]
stocks
```

### **3.2 Ordenando com base na Dívida Bruta sobre o Patrimônio Líquido(divbpatr)**

```python
stocks = df[(df.patrliq > 0) & (df.divbpatr <= 0.6)].sort_values('divbpatr', ascending = True)
stocks
```

### **3.3 Filtrando as 50 Primeiras Empresas `head(50)`**

```python
stocks = stocks.head(50)
stocks
```

### **3.3 Definindo a Função `is_traded_recently`**

```python
def is_traded_recently(acao):
    try:
        stock = yf.Ticker(acao)  # Obtém informações da ação usando o Ticker do yfinance
        historical_data = stock.history(period ='1mo')  # Obtém o histórico de negociação dos últimos 30 dias

        last_trading_day = historical_data.index[-1].date()  # Obtém a data do último dia de negociação

        today = datetime.now().date()  # Obtém a data atual
        window_start = today - timedelta(days = 5)  # Calcula o início da janela de tempo de 5 dias antes da data atual

        # Retorna True se o último dia de negociação estiver dentro da janela de tempo definida
        return window_start <= last_trading_day <= today
    
    except Exception:
        return False  # Retorna False em caso de erro
```

```python
is_traded_recently('PETR4.SA')
```

### **3.4** Por conta da biblioteca `yfinance`, precisamos **adicionar ".SA" ao final dos tickers**. Se usarmos `stocks.index`, conseguimos analisar apenas os tickers. Com isso, podemos **concatenar o sufixo** ` + '.SA'` e **criar uma nova coluna** ('acao')

```python
stocks['acao'] = stocks.index + '.SA'
stocks
```

### **3.5** **Aplicando a Função** `is_traded_recently` na coluna criada ('acao'), e atribuindo o resultado em uma **nova coluna** ('negociada')

```python
stocks['negociada'] = stocks['acao'].apply(is_traded_recently)
stocks
```

### **3.6 Filtrando as Empresas que São Negociadas `negociada == True`**

```python
stocks = stocks[stocks['negociada'] == True]
stocks
```

### **3.7** **Ordenando** com base no **ROE (roe)** e **selecionando** as **10 Primeiras Empresas** 

```python
stocks = stocks.sort_values('roe', ascending = False).head(10)
stocks
```

### **3.8 Análise Gráfica**

```python
plt.figure(figsize=(10,6))
plt.bar(stocks.tail(10).index, stocks.tail(10).roe)
plt.xticks(rotation = 45)
plt.title("Screening de Ações Dívida Bruta/Patrimônio Líquido e ROE")
```
